```python
experiments = ['square', 'random', 'chirp']
runs = [1, 2, 3]
df_real_dict = {}

for experiment in experiments:
    file_name = f'{experiment}_20251017'
    for run in runs:
        try:
            file_name_real = f'{file_name}_run{run}.csv'
            df_real = pd.read_csv(os.path.join('../data/train/', file_name_real))
            # Store dynamically
            df_real_dict[f'{experiment}_run{run}'] = df_real
        except Exception as e:
            print(e)
            continue

experiments = ["melon"] # Choose: ['square', 'random', 'chirp']
runs = [1, 2, 3]

for experiment in experiments:
    file_name = f'{experiment}_20251017'
    for run in runs:
        try:
            file_name_real = f'{file_name}_run{run}.csv'
            df_real = pd.read_csv(os.path.join('../data/test/', file_name_real))
            # Store dynamically
            df_real_dict[f'{experiment}_run{run}'] = df_real
        except Exception as e:
            print(e)
            continue
```
